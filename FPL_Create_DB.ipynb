## Fantasy Premier League


```python
import pandas as pd
import sqlite3
import requests
from io import StringIO
```

```python
# Base URL
base_url = "https://raw.githubusercontent.com/olbauday/FPL-Core-Insights/main/data/2025-2026/"

# Download all 4 tables
gameweek_summaries = pd.read_csv(base_url + "gameweek_summaries.csv")
players = pd.read_csv(base_url + "players.csv")
playerstats = pd.read_csv(base_url + "playerstats.csv")
teams = pd.read_csv(base_url + "teams.csv")
```

```python
playerstats
```

```python
print("Players columns:", players.columns.tolist())
print("Teams columns:", teams.columns.tolist())
print("Player stats columns:", playerstats.columns.tolist())
print("Gameweek summaries columns:", gameweek_summaries.columns.tolist())
```

```python
# The relationships between tables:
# 1. players.player_id <-> playerstats.id
# 2. players.player_id <-> playerstats (player stats table)
# 3. teams.id <-> players.team_code (teams the players play for)

# Save to SQLite database with these relationships in mind
conn = sqlite3.connect('fpl.db')

players.to_sql('players', conn, if_exists='replace', index=False)
playerstats.to_sql('playerstats', conn, if_exists='replace', index=False)
teams.to_sql('teams', conn, if_exists='replace', index=False)
gameweek_summaries.to_sql('gameweek_summaries', conn, if_exists='replace', index=False)

print("✅ Database created!")
print("\nKey relationships:")
print("- players.player_id links to playerstats.id")
print("- players.team_code links to teams.code")
print("- teams.id is the team identifier")
```
